```python
#Line plot 2
year = [1990, 2000, 2010, 2020]
pop = [2, 3, 5, 9]
plt.plot(year,pop)

#Axes
plt.xlabel('year')
plt.ylabel('population')
#Title
plt.title('idek man')

#Edit scale, what to show for the scale value
plt.yticks([0,2,4,6,8,10],['0B','2B','4B','6B','8B','10B'])
plt.show()
```

```python
#Line plot
year = [1990, 2000, 2010, 2020]
pop = [256, 348, 589, 987]
np_pop = np.array([20,40,60,80])
#S sets the size of the scatter dots
#c is the colour
#alpha is the opacity
plt.scatter(year,pop, s= np_pop, c='pink', alpha = 0.5)

#Axes
plt.xlabel('year')
plt.ylabel('population')
#Title
plt.title('idek man')
plt.show()
```

```python
#Line plot
year = [1990, 2000, 2010, 2020]
pop = [256, 348, 589, 987]
plt.scatter(year,pop)

#Axes
plt.xlabel('year')
plt.ylabel('population')
#Title
plt.title('idek man')

#Adds text next to the datapoint
plt.text(1990,256,"lmao")
plt.text(2000,348,"why tho")

#Add grid lines
plt.grid(True)

plt.show()
```
